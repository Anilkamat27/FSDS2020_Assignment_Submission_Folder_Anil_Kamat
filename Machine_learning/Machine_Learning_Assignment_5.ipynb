**2.Describe quantitative and qualitative data in depth. Make a distinction between the two.**

Quantitative Data:

It consists of numerical values.
Measured by numbers (e.g., profit, quantity sold, height, weight).
Two types:
Discrete (countable, like the number of cars or students).
Continuous (measurable within a range, like height or time).

Qualitative Data:

It cannot be expressed in numbers.
Describes categories or groups.
Two types:
Structured (expressed as numbers or words, but no mathematical operations can be performed).
Unstructured (no proper format, includes textual data, sounds, images, videos).

**9.i. What is the IQR? What criteria are used to assess it?**

Q1 is the first quartile of the data, i.e., to say 25% of the data lies between minimum and Q1. Q3 is the third quartile of the data, i.e., to say 75% of the data lies between minimum and Q3. The difference between Q3 and Q1 is called the Inter-Quartile Range or IQR.

**ii. Describe the various components of a box plot in detail? When will the lower whisker surpass the upper whisker in length? How can box plots be used to identify outliers?**

arious Components of a Box Plot:

Minimum: The smallest value in the dataset.

Maximum: The largest value in the dataset.

Range: The difference between the maximum and minimum values, representing the dataset's spread.

Median (Q2): The middle point of the data when ordered, also known as the second quartile.

Q1 (First Quartile): The median of the lower half of the data, representing the 25th percentile.

Q3 (Third Quartile): The median of the upper half of the data, representing the 75th percentile.

Inter-Quartile Range (IQR): The difference between Q3 and Q1.

Lower Whisker vs. Upper Whisker Length:

When the data is left-skewed (negatively skewed), the lower whisker will be longer than the upper whisker.

Identifying Outliers:

The box plot helps identify outliers using the Inter-Quartile Range (IQR).
Decision Range for Outliers:

Lower Bound: (Q1 - 1.5 * IQR)
Upper Bound: (Q3 + 1.5 * IQR)

Any data point outside this range is considered an outlier.

Comparison: Data with Nominal and Ordinal Values:

Nominal Data: Represents categories or labels with no inherent order.
Examples: Colors, gender.No numerical significance; only labels.

Ordinal Data:Represents categories with a meaningful order.
Examples: Education levels (e.g., high school, bachelor's, master's).
Order matters, but the intervals between values are not consistent.

Comparison: Histogram and Box Plot:

Histogram:Displays the distribution of a continuous dataset.
Provides insights into data frequency and distribution shape.
Suitable for large datasets, less detail on central tendency and spread.

Box Plot (Box-and-Whisker Plot):Displays the distribution of a dataset using quartiles.
Highlights central tendency, spread, and identifies outliers.
Compact visualization, useful for comparing multiple datasets.
Comparison: The Average and Median:

Average (Mean):
Calculated by summing all values and dividing by the number of observations.
Sensitive to extreme values (outliers).
Represents the balance point of the dataset.

Median:
The middle value in a sorted dataset.
Less affected by extreme values.
Represents the central position of the dataset.
